```python
from langchain_openai import ChatOpenAI
from google.auth import default
from google.auth.transport.requests import Request

# 1. Update your project details
project_id = "sbx-31371-79b51c7781a2"
location = "europe-west1" 
# Use the correct model string for the OpenAI adapter
model_id = "google/claude-3-5-sonnet-v2@20241022" 

# 2. Get Credentials (Same as before)
credentials, _ = default(scopes=["https://www.googleapis.com/auth/cloud-platform"])
credentials.refresh(Request())

# 3. Corrected LangChain Initialization
llm = ChatOpenAI(
    model=model_id,
    openai_api_key=credentials.token,
    # Use the regional 'endpoints/openapi' base URL
    openai_api_base=f"https://{location}-aiplatform.googleapis.com/v1/projects/{project_id}/locations/{location}/endpoints/openapi",
)

# 4. Usage
response = llm.invoke("What are the benefits of using a multi-agent supervisor pattern?")
print(response.content)

```
